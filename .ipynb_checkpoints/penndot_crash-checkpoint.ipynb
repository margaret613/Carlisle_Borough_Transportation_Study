# Transforming and Visualizing PennDOT Crash Data

**Assigment: Create a CSV file, organized by PennDOT's Municipal ID Number, which contains variables for the total number of cyclists killed or suspected of serious injury from crashes with automobiles over the most recent 10-year period in which data is available. Sort the values by the total number of cyclists killed or seriously injured from greatest to least.**
- Columns use:
+ BICYCLE_DEATH_COUNT: Total amount of Bicyclist Fatalities
+ BICYCLE_SUSP_SERIOUS_INJ_COUNT: Total amount of Bicyclist Suspected Serious Injuries
+ AUTOMOBILE_COUNT: Total amount of Automobiles Involved
+ BICYCLE_COUNT: Total amount of Bicycles involved 
+ MUNICIPALITY: Municipality Code

## I. Data Cleaning & Data Manipulation

```python
import pandas as pd
```

```python
# Create an empty dataframe
df_crash = pd.DataFrame()

# Loop through the years 2011 to 2021
for year in range(2012, 2022):
    
    # Read the CSV file for the current year
    crash_year = pd.read_csv("./data/Statewide_{}/CRASH_{}_Statewide.csv"\
                             .format(year, year), low_memory=False)
    
    # Copy crash_year, store df_year
    df_year = crash_year.copy()
    
    # Keep necessary columns
    df_year = df_year[['MUNICIPALITY', 'BICYCLE_DEATH_COUNT', 'BICYCLE_SUSP_SERIOUS_INJ_COUNT', 'AUTOMOBILE_COUNT',\
                       'BICYCLE_COUNT']]
    
    # Filtering rows where BICYCLE_COUNT is not equal to 0
    df_year = df_year[df_year['BICYCLE_COUNT'] != 0]
    
    # Filtering rows where AUTOMOBILE_COUNT is not equal to 0
    df_year = df_year[df_year['AUTOMOBILE_COUNT'] != 0]
    
    # Grouping by MUNICIPALITY and calculating the total of each column 
    df_year = df_year.groupby('MUNICIPALITY').sum()
    
    # Append the current year's dataframe to the final DataFrame
    df_crash = pd.concat([df_crash, df_year])
```

```python
# Copy df_final, store df_final
df_final = df_crash.copy()
```

```python
# Calculating the total of each column 
df_final = df_final.groupby('MUNICIPALITY').sum()

# Create a new column calculating the total number of cyclists killed or suspected of 
# serious injury from crashes with automobiles over the most recent 10-year period
df_final['cyclists_killed_inj_count'] = df_final['BICYCLE_DEATH_COUNT'] + df_final['BICYCLE_SUSP_SERIOUS_INJ_COUNT']

# Drop unnecessary column (AUTOMOBILE_COUNT and BICYCLE_COUNT)
df_final = df_final.drop(columns = ['AUTOMOBILE_COUNT', 'BICYCLE_COUNT'], axis = 1)

# Resetting the index and making 'MUNICIPALITY' a column
df_final = df_final.reset_index()

# Reorganize the columns
df_final = df_final[['MUNICIPALITY', 'cyclists_killed_inj_count', 'BICYCLE_SUSP_SERIOUS_INJ_COUNT',\
                     'BICYCLE_DEATH_COUNT']]

# Sort the values by the total number of cyclists killed or seriously injured from greatest to least
df_final = df_final.sort_values('cyclists_killed_inj_count', ascending = False)

# Rename the columns
df_final.rename(columns={'MUNICIPALITY': 'Penn Dot Municipal ID number',
                         'cyclists_killed_inj_count': 'Total Number of Cyclists killed or suspected of serious injury from crashes with automobiles',\
                         'BICYCLE_SUSP_SERIOUS_INJ_COUNT': 'Total cyclists suspected of serious injury from crashes with automobiles over 10 year period',\
                         'BICYCLE_DEATH_COUNT': 'Total cyclists killed from crashes with automobiles over 10 year period'}, inplace=True)

# Reset the index and rename the index column
df_final.reset_index(drop=True, inplace=True)
df_final.rename_axis('Index', inplace=True)
```

```python
# View dataframe
df_final.head(10)
```

```python
# Export dataframe to csv file
df_final.to_csv('penndot_crash.csv',index=True)
```

## II. Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly
```

```python
# Only get top 10 municipalities in PA 
df = df_final.head(10).copy()
```

```python
# Create a stacked bar chart
df.plot(x='Penn Dot Municipal ID number',\
        y=['Total cyclists suspected of serious injury from crashes with automobiles over 10 year period',\
           'Total cyclists killed from crashes with automobiles over 10 year period'], kind='barh', stacked=True)

# Add labels and title
plt.xlabel('Total Number of Cyclists')
plt.ylabel('Penn Dot Municipal ID number')
plt.title('Top 10 Municipalties in Pennsylvania: Cyclist Fatalities and Serious Injuries from Automobile Crashes (2012-2021)')

# Show the plot
plt.show()
```

```python
# Get the number of data points
n_points = len(df)

# Create bars for the chart
bar1 = df['Total cyclists killed from crashes with automobiles over 10 year period']
bar2 = df['Total cyclists suspected of serious injury from crashes with automobiles over 10 year period']
bar3 = df['Total Number of Cyclists killed or suspected of serious injury from crashes with automobiles']

# Set up the plot
x_max = 2 * np.pi
x_coords = np.linspace(0, x_max, n_points, endpoint = False)
width = x_max / n_points

# Define the inner radius
inner_radius = 1

# Create the figure and axis
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

# Plot the bars
ax.bar(
    x_coords,
    bar3,
    width = width,
    bottom = inner_radius, 
    color = "lightblue",
    edgecolor = "white",
    linewidth = 1
)
ax.bar(
    x_coords,
    bar2,
    width = width,
    bottom = inner_radius,
    color = "grey",
    edgecolor = "white",
    linewidth = 1
)
bars = ax.bar(
    x_coords,
    bar1,
    width = width,
    bottom = inner_radius,
    color = "pink",
    edgecolor = "white",
    linewidth = 1
)

# Set the axis off
plt.axis("off")

# Display the plot
plt.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Barpolar(
    r = list(df['Total Number of Cyclists killed or suspected of serious injury from crashes with automobiles']),
    theta =list(df['Penn Dot Municipal ID number']),
    name = 'Killed or Suspected',
    marker_color ='rgb(46,109,255)',
    marker_line_color ="black",
    hoverinfo = ['all'],
    opacity = 0.7   
))

fig.add_trace(go.Barpolar(
    r = list(df['Total cyclists suspected of serious injury from crashes with automobiles over 10 year period']),
    theta = list(df['Penn Dot Municipal ID number']),
    name = 'Suspected',
    marker_color ='rgb(139,222,52)',
    marker_line_color ="black",
    hoverinfo = ['all'],
    opacity = 0.7
))

fig.add_trace(go.Barpolar(    
    r = list(df['Total cyclists killed from crashes with automobiles over 10 year period']),
    theta=list(df['Penn Dot Municipal ID number']),
    name = 'Killed',
    marker_color = 'rgb(245,94,21)', 
    marker_line_color = "black",
    hoverinfo = ['all'],
    opacity = 0.7
))
fig.update_layout(
    title = 'Cyclist Fatalities and Serious Injuries from Crashes with Automobiles in PA (10-Year Data)',
    font_size = 12,
    legend_font_size = 15,
    polar_angularaxis_rotation = 90,
    width = 900,
    height = 900,
    
    polar = dict(
              bgcolor = "rgb(223, 223, 223)",
              angularaxis = 
                dict(
                    linewidth = 3,
                    showline = True,
                    linecolor = 'black'
                    ),
          radialaxis = 
                dict(        
                    showline = True,
                    linewidth = 2,
                    gridcolor = "white",
                    gridwidth = 2,
                    )
                ),

            )
#fig.show()
plotly.offline.plot(fig, filename='penndot_crash.html')
```
