# Generate scatter plots

Assignment: Generate scatter plots using the list of variables from the df_pa_crash.
List of variables:
- BNA Score
- POPULATION  
- LAND_AREA 
- BIKE_TO_WORK_EST: The Bike to work estimate is the American Community Survey's 5-year estimate for the number of people who commute to work on a bicycle. It is queried from the 2021 dataset for each of the county subdivisions in Pennsylvania. 
- BICYCLE_BY_AUTO_COUNT: The number of cyclists' deaths and serious injuries 2017-2021.
- BICYCLE_DEATH_BY_AUTO_COUNT
    + BY_AUTO_COUNT should refer to cyclists deaths in crashes with automobiles. 
- BICYCLE_SUSP_SERIOUS_INJ_BY_AUTO_COUNT 
    + BY_AUTO_COUNT should refer to cyclists seriously injured in crashes with automobiles. 

```python
# Import packages 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import data
df_pa_crash = pd.read_csv('/Users/margaret06/Documents/GitHub/Carlisle_Borough_Transportation_Study/data/df_pa_crash.csv')

# Clean datasets
df_pa_crash = df_pa_crash.drop(columns = ['Unnamed: 0'])
```

```python
# Check the data types
df_pa_crash.dtypes
```

```python
# View the dataframe
df_pa_crash.head()
```

**Are municipalities with higher populations more likely to have higher BNA Scores, or is there no clear relationship?**

```python
# Create scatter plot
df_pa_crash.plot.scatter(x='BNA Score', y='POPULATION', title='Scatter Plot of BNA Score vs. Population') # pandas
plt.xlabel('BNA Score')
plt.ylabel('Population')

# Show the plot
plt.grid(True)
plt.show()
```

**Does a municipality with larger land areas tend to have higher or lower BNA Scores?**

```python
# Create scatter plot
df_pa_crash.plot.scatter(x='BNA Score', y='LAND_AREA', title='Scatter Plot of BNA Score vs. Land Area') # pandas
plt.xlabel('BNA Score')
plt.ylabel('Land Area')

# Show the plot
plt.grid(True)
plt.show()
```

**Is there a connection between the BNA Score and the number of people commuting to work on bicycles?**

```python
# Create scatter plot
df_pa_crash.plot.scatter(x='BNA Score', y='BIKE_TO_WORK_EST', title='Scatter Plot of BNA Score vs. The number of people who commute to work on a bicycle') # pandas
plt.xlabel('BNA Score')
plt.ylabel('The number of people who commute to work on a bicycle')

# Show the plot
plt.grid(True)
plt.show()
```

**Does the total count of cyclist-involved crashes (BICYCLE_BY_AUTO_COUNT) vary with the BNA Score?**

```python
# Create scatter plot
df_pa_crash.plot.scatter(x='BNA Score', y='BICYCLE_BY_AUTO_COUNT', title='Scatter Plot of BNA Score vs. The number of cyclists deaths and serious injuries 2017-2021') # pandas
plt.xlabel('BNA Score')
plt.ylabel('The number of cyclists deaths and serious injuries 2017-2021')

# Show the plot
plt.grid(True)
plt.show()
```

**Is there a correlation between the total count of cyclists deaths in crashes with automobiles and the BNA score?**

```python
# Create scatter plot with x and y swapped
df_pa_crash.plot.scatter(x='BNA Score', y='BICYCLE_DEATH_BY_AUTO_COUNT', title='BNA Score vs. Cycling Deaths 2017-2021: A Scatter Plot') # pandas

# Labels
plt.xlabel('BNA')
plt.ylabel('The number of cyclists deaths 2017-2021')

# Show the plot
plt.grid(True)
plt.show()
```

**Is there a correlation between the total count of cyclists seriously injured in crashes with automobiles and the BNA score?**

```python
# Create scatter plot with x and y swapped
df_pa_crash.plot.scatter(x='BNA Score', y='BICYCLE_SUSP_SERIOUS_INJ_BY_AUTO_COUNT', title='BNA Score vs. Cycling Seriously Injured 2017-2021: A Scatter Plot') # pandas

# Labels
plt.xlabel('BNA')
plt.ylabel('The number of cyclists seriously injured 2017-2021')

# Show the plot
plt.grid(True)
plt.show()
```

**Are areas with a higher population more likely to have a higher total count of cyclist-involved crashes?**

```python
# Create scatter plot with x and y swapped
df_pa_crash.plot.scatter(x='POPULATION', y='BICYCLE_BY_AUTO_COUNT', title="Population vs. Cycling Casualties 2017-2021: Scatter Plot") # pandas

# Labels
plt.xlabel('Population')
plt.ylabel('The number of cyclists deaths and serious injuries 2017-2021')

# Show the plot
plt.grid(True)
plt.show()
```
