List of variables:

- BICYCLE_BY_AUTO_COUNT (cyclists deaths and serious injuries 2012-2021)

```python
# Import packages 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import data
df_pa_crash = pd.read_csv('/Users/margaret06/Documents/GitHub/Carlisle_Borough_Transportation_Study/data/df_pa_crash.csv')

# Clean datasets
df_pa_crash = df_pa_crash.drop(columns = ['Unnamed: 0'])
```

```python
# Check the data types
df_pa_crash.dtypes
```

```python
# View the dataframe
df_pa_crash.head()
```

**Are municipalities with higher populations more likely to have higher BNA Scores, or is there no clear relationship?**

```python
# Create scatter plot
df_pa_crash.plot.scatter(x='BNA Score', y='POPULATION', title='Scatter Plot of BNA Score vs. Population') # pandas
plt.xlabel('BNA Score')
plt.ylabel('Population')

# Fit a linear regression line
slope, intercept = np.polyfit(df_pa_crash['BNA Score'], df_pa_crash['POPULATION'], 1)

# Create the line equation
line = slope * df_pa_crash['BNA Score'] + intercept

# Plot the line of best fit
plt.plot(df_pa_crash['BNA Score'], line, color='red', linestyle='--', label='Line of Best Fit')

# Display the legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Create scatter plot
df_pa_crash.plot.scatter(x='BNA Score', y='LAND_AREA', title='Scatter Plot of BNA Score vs. Land Area') # pandas
plt.xlabel('BNA Score')
plt.ylabel('Land Area')

# Fit a linear regression line
slope, intercept = np.polyfit(df_pa_crash['BNA Score'], df_pa_crash['LAND_AREA'], 1)

# Create the line equation
line = slope * df_pa_crash['BNA Score'] + intercept

# Plot the line of best fit
plt.plot(df_pa_crash['BNA Score'], line, color='red', linestyle='--', label='Line of Best Fit')

# Display the legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
